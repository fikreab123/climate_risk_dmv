Maryland Climate Dataset
This dataset shows climate risk scores for each county in Maryland.
It includes population and ratings that describe how likely each county is to face things like floods, heat, and other disasters.

```python
climate_md = pd.read_csv("Maryland Climate Dataset.csv")

climate_md.head()
```

```python
# fix column names
climate_md = climate_md.rename(columns={
    'County': 'COUNTY',
    'Risk Score': 'RISK_SCORE',
    'Risk Rating': 'RISK_RATNG'})
```

```python
#Slice only the important columns 
climate_md = climate_md[[
    'COUNTY', 'STATE', 'POPULATION',
    'RISK_SCORE', 'RISK_RATNG',
    'EAL_SCORE', 'EAL_RATNG']]
```

City-to-County Dataset
This file helps me find out what county each city is in.
I need it because my bank data only has city names, and my climate data uses counties.
This dataset lets me connect the two.

```python
city_county = pd.read_csv("City-to-County Dataset.csv")

city_county.head()
```

```python
# Clean city names
df_md['CITY'] = df_md['CITY'].str.strip().str.title()
city_county['city'] = city_county['city'].str.strip().str.title()
city_county['state_name'] = city_county['state_name'].str.strip().str.title()

# Keep only Maryland cities
city_county_md = city_county[city_county['state_name'] == 'Maryland']

# Merge to add the correct county name
df_md = df_md.merge(
    city_county_md[['city', 'county_name']],
    how='left',
    left_on='CITY',
    right_on='city')
```

```python
df_md.head()
```

```python
df_md[df_md['county_name'].isnull()]['CITY'].unique()
```

```python
# Manually add missing county names 
df_md.loc[df_md['CITY'] == 'Upperco', 'county_name'] = 'Baltimore'
df_md.loc[df_md['CITY'] == 'Hunt Valley', 'county_name'] = 'Baltimore'

# Rename
df_md = df_md.rename(columns={
    'NAMEFULL': 'BANK_NAME',
    'ADDRESBR': 'STREET_ADDRESS',
    'ZIP': 'ZIP_CODE',
    'DEPSUMBR': 'TOTAL_DEPOSITS',
    'NAMEBR': 'BRANCH_NAME',
    'SIMS_ESTABLISHED_DATE': 'ESTABLISHED_DATE',
    'SIMS_LATITUDE': 'LATITUDE',
    'SIMS_LONGITUDE': 'LONGITUDE'
})
```

Merging Bank and Climate Data
I will now combine the bank data with the climate data using the COUNTY column.
This will help each bank branch get its matching climate risk score.

```python
climate_md.columns
```

```python
df_md.columns
```

```python
# rename column
df_md = df_md.rename(columns={'county_name': 'COUNTY'})

# fix Nottingham county
df_md.loc[df_md['CITY'] == 'Nottingham', 'COUNTY'] = 'Baltimore'

# drop extra city column
df_md = df_md.drop(columns=['city'])

# drop old risk columns
df_md = df_md.drop(columns=['RISK_SCORE', 'RISK_RATNG'], errors='ignore')
```

```python
# Merge climate data into bank data using COUNTY
df_md = df_md.merge(climate_md[['COUNTY', 'RISK_SCORE', 'RISK_RATNG']], on='COUNTY', how='left')
```

```python
df_md.duplicated().sum()
```

```python
# rename all columns to clean names
df_md = df_md.rename(columns={
    'CERT': 'cert',
    'BANK_NAME': 'bank_name',
    'STREET_ADDRESS': 'address',
    'CITY': 'city',
    'STNAME': 'state',
    'ZIP_CODE': 'zip',
    'TOTAL_DEPOSITS': 'deposit',
    'BRANCH_NAME': 'branch_name',
    'ESTABLISHED_DATE': 'established_date',
    'LATITUDE': 'latitude',
    'LONGITUDE': 'longitude',
    'COUNTY': 'county',
    'RISK_SCORE': 'risk_score',
    'RISK_RATNG': 'risk_rating'
})

# fix date
df_md['established_date'] = pd.to_datetime(df_md['established_date'])
```

```python
df_md.isnull().sum()
```

```python
# keep only highest risk score per cert + address
df_md = df_md.sort_values('risk_score', ascending=False)
df_md = df_md.drop_duplicates(subset=['cert', 'address'], keep='first')
```
