Virginia Climate Dataset
This dataset shows climate risk scores for each county in VA.
It includes population and ratings that describe how likely each county is to face things like floods, heat, and other disasters.

```python
climate_va = pd.read_csv("Virginia Climate Dataset.csv")
list(climate_va.columns)

```

```python
# Keep only useful columns
climate_va = climate_va[['COUNTY', 'STATE', 'COUNTYFIPS', 'POPULATION',
                         'RISK_SCORE', 'RISK_RATNG', 'EAL_SCORE', 'EAL_RATNG']]

climate_va.head()
```

City-to-County Dataset
This file helps me find out what county each city is in.
I need it because my bank data only has city names, and my climate data uses counties.
This dataset lets me connect the two.

```python
city_county = pd.read_csv("City-to-County Dataset.csv")

city_county.head()
```

```python
city_county.columns
```

```python
# Keep only needed columns and filter for Virginia
city_county = city_county[['city', 'county_name', 'state_name']]
city_county = city_county[city_county['state_name'] == 'Virginia']

# Clean city names in both datasets
df['CITY'] = df['CITY'].str.strip().str.upper()
city_county['city'] = city_county['city'].str.strip().str.upper()
```

```python
# Merge Virginia bank data with city-to-county dataset
df = df.merge(city_county,
              left_on=['CITY', 'STNAME'],
              right_on=['city', 'state_name'],
              how='left')

# Drop extra columns from merge
df = df.drop(columns=['city', 'state_name'])
```

```python
df.isna().sum()
```

```python
df[df['county_name'].isna()]['CITY'].unique()
```

```python
# Manual fixes for unmatched cities
df.loc[df['CITY'] == 'MC KENNEY', 'county_name'] = 'DINWIDDIE'
df.loc[df['CITY'] == 'BLUE GRASS', 'county_name'] = 'HIGHLAND'
df.loc[df['CITY'] == 'CARSON', 'county_name'] = 'DINWIDDIE'
df.loc[df['CITY'] == 'RICHMOND CITY', 'county_name'] = 'RICHMOND CITY'
```

```python
# Make names match before merge
df['county_norm'] = df['county_name'].str.upper().str.replace(' COUNTY','').str.replace(' CITY','').str.strip()
climate_va['COUNTY_NORM'] = climate_va['COUNTY'].str.upper().str.replace(' COUNTY','').str.replace(' CITY','').str.strip()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df['STNAME'].value_counts()
```

```python
# Merge Virginia bank data with Virginia climate data
df_va = df.merge(climate_va, left_on='county_norm', right_on='COUNTY_NORM', how='left')

```

```python
df_va.isnull().sum()
```

```python
df_va.duplicated().sum()
```

```python
df_va.columns
```

```python
df_va = df_va.drop(columns=['county_norm', 'COUNTY_NORM'])
```

```python
df_va.head()
```

```python
# Rename columns for Virginia
df_va = df_va.rename(columns={
    'CERT': 'cert',
    'NAMEFULL': 'bank_name',
    'ADDRESBR': 'address',
    'CITY': 'city',
    'STNAME': 'state',
    'ZIP': 'zip',
    'DEPSUMBR': 'deposit',
    'NAMEBR': 'branch_name',
    'SIMS_ESTABLISHED_DATE': 'established_date',
    'SIMS_LATITUDE': 'latitude',
    'SIMS_LONGITUDE': 'longitude',
    'county_name': 'county',
    'RISK_SCORE': 'risk_score',
    'RISK_RATNG': 'risk_rating'
})

# Keep only needed columns
df_va = df_va[[
    'cert', 'bank_name', 'address', 'city', 'state', 'zip', 'deposit',
    'branch_name', 'established_date', 'latitude', 'longitude',
    'county', 'risk_score', 'risk_rating'
]]
```

```python
df_va.columns
```

```python
df_va.head()
```

```python
df_va = df_va.sort_values('risk_score', ascending=False)
df_va = df_va.drop_duplicates(subset=['cert', 'address'], keep='first')
```

```python
df_va.to_csv("va_clean.csv", index=False)
```
