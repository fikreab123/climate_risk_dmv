Wasington DC Climate Dataset
This dataset shows climate risk scores for each county in Wasington DC.
It includes population and ratings that describe how likely each county is to face things like floods, heat, and other disasters.

```python
climate_dc = pd.read_csv("DC Climate Dataset.csv")
climate_dc.head()
```

```python
climate_dc.columns.tolist()
```

```python
# fix column names
climate_dc = climate_dc.rename(columns={
    'COUNTY': 'county',
    'STATE': 'state',
    'POPULATION': 'population',
    'RISK_SCORE': 'risk_score',
    'RISK_RATNG': 'risk_rating'
})

# keep only needed columns
climate_dc = climate_dc[['county', 'state', 'population', 'risk_score', 'risk_rating']]


climate_dc.head()
```

```python
# add county column
df_dc['county'] = 'District of Columbia'

# merge with DC climate
dc_merged = df_dc.merge(climate_dc, on='county', how='left')

# quick look
dc_merged.head()
```

```python
print(dc_merged.columns.tolist())
```

```python
# rename all columns to clean names
df_dc = dc_merged.rename(columns={
    "CERT": "cert",
    "NAMEFULL": "bank_name",
    "ADDRESBR": "address",
    "CITY": "city",
    "STNAME": "state",
    "ZIP": "zip",
    "DEPSUMBR": "deposit",
    "NAMEBR": "branch_name",
    "SIMS_ESTABLISHED_DATE": "established_date",
    "SIMS_LATITUDE": "latitude",
    "SIMS_LONGITUDE": "longitude",
    "county_name": "county",           
    "RISK_SCORE": "risk_score",       
    "RISK_RATNG": "risk_rating"
})

# keep column order
df_dc = df_dc[[
    "cert","bank_name","address","city","state","zip","deposit",
    "branch_name","established_date","latitude","longitude",
    "county","risk_score","risk_rating"
]]
```
